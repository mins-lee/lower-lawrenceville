```python
import json
import requests
import pandas as pd
```

```python
links = []
with open('links.txt', 'r') as f:
    for value in f:
        links.append(value.rstrip())
        
titles = []
with open('titles.txt', 'r') as f:
    for value in f:
        titles.append(value.rstrip())
        
url_lead = 'https://data.wprdc.org/api/3/action/datastore_search?resource_id='
url_cond = '&limit=5'
headers = {'Content-Type': 'application/json'}

Yr = []
Mon = [] 
Field = [] 
Type = [] 
Field_Count = []
failed = []
        
for i in range(0,len(links)):
    resid = links[i].split('/')[-1]
    yr = titles[i].split()[-1]
    mon = titles[i].split()[-2]
    
    url = url_lead + resid + url_cond
    resp = requests.get(url, headers = headers)
    
    if resp.status_code == 200:
        data = json.loads(resp.content.decode('utf-8'))

        #print(i)
        counter = 0
        for d in data['result']['fields']:
            Yr.append(yr)
            Mon.append(mon)
            Field.append(d['id'].lower())
            Type.append(d['type'])
            Field_Count.append(counter)
            counter += 1
        
    else:
        failed.append([mon+' '+yr,resid])

df = pd.DataFrame(list(zip(Yr, Mon,Field,Type,Field_Count)),
              columns=['Yr', 'Mon','Field','Type','Field_count'])
```

# Run API call over all API resource ids

```python
import json
import requests
import pandas as pd
import numpy as np
```

```python
links = []
with open('links.txt', 'r') as f:
    for value in f:
        links.append(value.rstrip())
        
titles = []
with open('titles.txt', 'r') as f:
    for value in f:
        titles.append(value.rstrip())
```

```python
# Build urls
url_lead = 'https://data.wprdc.org/api/3/action/datastore_search?resource_id='
url_cond = '&limit=5000'
headers = {'Content-Type': 'application/json'}
```

```python
# Set a dict to rename columns to create unified names to join on
# if contains, then full replace this:
col_dict = {'address':'address',
    'contractor':'contractor',
    'value':'cost',
    'cost':'cost',
    'date':'issued',
    'neigh':'neighborhood',
    'owner':'owner',
    'parcel':'parcel',
    'permit':'permit',
    'ward':'ward',
    'struct':'structure_type',
    'type of work':'work_type',
    'typeofwork':'typeofwork_description',
    'work description':'work description',
    'workdescription':'workdescription',
    '_id':'id'
    }

conditions = '|'.join(list(col_dict.keys()))
```

```python
# Set lists to capture possible API failures
failed_API = [] # If dataset does not have API data
no_records = [] # if API call works, but there are no records
failed_rename = [] # if the rename dictionary doesn't fully capture columns
df_list = [] # Successful API data
```

```python
for i in range(0,len(links)):
    resid = links[i].split('/')[-1]
    yr = titles[i].split()[-1]
    mon = titles[i].split()[-2]

    url = url_lead + resid + url_cond
    
    r = requests.get(url, headers = headers)

    if r.status_code == 200:
        data = json.loads(r.content.decode('utf-8'))

        df = pd.json_normalize(data, ['result', 'records'])
        
        regex_arr = df.columns.to_series().str.lower().str.contains(conditions, regex=True)

        if regex_arr.all():
            regex_columns = df.columns.to_series().str.lower()[regex_arr]
            for v in regex_columns:
                for key in col_dict.keys():
                    if key in v:
                        regex_columns.replace(v,col_dict[key], inplace=True)

            df.columns = regex_columns
            df['year'] = yr
            df['month'] = mon
            if len(df)==0:
                no_records.append([mon+' '+yr,resid])
            else:
                df_list.append(df)
            
        else:
            failed_rename.append([mon+' '+yr,resid])

    else:
        failed_API.append([mon+' '+yr,resid])
```
