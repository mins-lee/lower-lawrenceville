```python
geo = {}
```

```python
for i in range(0,len(data)):
    if data.loc[i,'address'] not in geo.keys():
        geo[data.loc[i,'address']] = {'lon':data.loc[i]['lon'], 'lat':data.loc[i]['lat'],
                                'geocoder_recognized_address':data.loc[i]['geocoder_recognized_address']}
```

## Use geocoding to fill out the duplicated addresses

```python
df = pd.read_csv('working.csv')
```

```python
df2013 = pd.read_csv('working2013.csv')
```

```python
df = pd.concat([df,df2013])
```

```python
df = df.reset_index()
```

```python
df = df.iloc[:,1:]
```

```python
df
```

```python
df['lon'] = df.address.apply(lambda x: geo[x]['lon'] if x in geo.keys() else np.nan)
```

```python
df['lat'] = df.address.apply(lambda x: geo[x]['lat'] if x in geo.keys() else np.nan)
```

```python
df['geocoder_recognized_address'] = df.address.apply(lambda x: geo[x]['geocoder_recognized_address'] if x in geo.keys() else np.nan)
```

```python
df.lon.isna().sum()
```

```python
df.columns
```

```python
df = df.loc[:,['permit', 'parcel', 'owner', 'contractor', 'neighborhood',  
       'ward', 'clean_ward', 'cost', 'issued', 'year', 'mn', 'address', 
          'geocoder_recognized_address', 'lon', 'lat', 'structure_type',  
       'work description', 'work_type']]
```

```python
df.to_csv('final_permit_data.csv', index=False)
```
